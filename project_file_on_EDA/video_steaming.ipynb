```python
x_axis = df['service_name']
label = df['category']
```

```python
subscribers_by_year_category = df.pivot_table(
    index='launch_year',
    columns='category',
    values='subscribers_millions',
    aggfunc='sum'
).fillna(0)

# --- Preparing Data for the Plot ---
# The x-axis is the years from our pivoted table index
x_axis = subscribers_by_year_category.index

# The y-values are the subscriber numbers for each category.
# We create a list of arrays, where each array is a column from our pivoted table.
arrays_to_plot = [subscribers_by_year_category[col] for col in subscribers_by_year_category.columns]

# The labels for our legend are the column names (the categories)
plot_labels = subscribers_by_year_category.columns

# --- Create the Stack Plot ---
plt.figure(figsize=(12, 8)) # Make the plot bigger for readability
plt.stackplot(x_axis, *arrays_to_plot, labels=plot_labels)

plt.legend(loc='upper left')
plt.title('Total Subscribers by Category Over Time')
plt.xlabel('Year')
plt.ylabel('Total Subscribers (Millions)')
plt.show()
```
